```python
def plot_all(dataset, init, title):
    x = np.linspace(init, 5000, len(dataset[0]['Fidelity History'][init:]))
    plt.plot(x, dataset[0]['Fidelity History'][init:], '.r')
    plt.plot(x, dataset[1]['Fidelity History'][init:], '.g')
    plt.plot(x, dataset[2]['Fidelity History'][init:], '.b')
    plt.plot(x, dataset[3]['Fidelity History'][init:], '.c')
    plt.plot(x, dataset[4]['Fidelity History'][init:], '.m')
    plt.xlabel('State Tomography')
    plt.ylabel('Fidelity')
    plt.title(title)
    plt.show()

    avg_ssdp = []
    avg_ssdp.append(dataset[0]['Fidelity History'][init:])
    avg_ssdp.append(dataset[1]['Fidelity History'][init:])
    avg_ssdp.append(dataset[2]['Fidelity History'][init:])
    avg_ssdp.append(dataset[3]['Fidelity History'][init:])
    avg_ssdp.append(dataset[4]['Fidelity History'][init:])
    avg_ssdp = np.array(avg_ssdp)

    mean_ssdp = avg_ssdp.mean(axis=0)
    std_ssdp = avg_ssdp.std(axis=0)
    print("5000 state tomography fidelity", round(mean_ssdp[-1], 3))
    plt.xlabel('State Tomography')
    plt.ylabel('Fidelity')
    plt.title(title)

    plt.plot(np.linspace(init, 5001, 5001-init), mean_ssdp, '.')
    plt.fill_between(np.linspace(init, 5001, 5001-init), mean_ssdp-std_ssdp, mean_ssdp+std_ssdp, alpha=0.5)
    plt.show()

    return round(mean_ssdp[-1], 3)
```
